```python
df_feature_eng = df.copy()
```

```python
df_feature_eng['ft_home_goals'] = df_feature_eng['ft_score'].str.extract('(^\d)', expand=True)
df_feature_eng['ft_away_goals'] = df_feature_eng['ft_score'].str.extract('(\d$)', expand=True)
df_feature_eng['ht_home_goals'] = df_feature_eng['ht_score'].str.extract('(^\d)', expand=True)
df_feature_eng['ht_away_goals'] = df_feature_eng['ht_score'].str.extract('(\d$)', expand=True)
```

```python
numeric_cols = ['ft_home_goals', 'ft_away_goals', 'ht_home_goals', 'ht_away_goals']

for col in numeric_cols:
    df_feature_eng[col] = pd.to_numeric(df_feature_eng[col])
```

```python
df_feature_eng.dtypes
```

```python
ft_conditions = [df_feature_eng['ft_home_goals'] > df_feature_eng['ft_away_goals'],
              df_feature_eng['ft_home_goals'] < df_feature_eng['ft_away_goals'],
              df_feature_eng['ft_home_goals'] == df_feature_eng['ft_away_goals']]

ft_result = [df_feature_eng['home_team'],
          df_feature_eng['away_team'],
          'Draw']

df_feature_eng['ft_result'] = np.select(ft_conditions, ft_result)
```

```python
df_feature_eng.head()
```

```python
ht_conditions = [df_feature_eng['ht_home_goals'] > df_feature_eng['ht_away_goals'],
              df_feature_eng['ht_home_goals'] < df_feature_eng['ht_away_goals'],
              df_feature_eng['ht_home_goals'] == df_feature_eng['ht_away_goals']]

ht_result = [df_feature_eng['home_team'],
          df_feature_eng['away_team'],
          'Draw']

df_feature_eng['ht_result'] = np.select(ht_conditions, ht_result)
```

```python
df_feature_eng['secondhalf_home_goals'] = df_feature_eng['ft_home_goals'] - df_feature_eng['firsthalf_home_goals']
df_feature_eng['secondhalf_away_goals'] = df_feature_eng['ft_away_goals'] - df_feature_eng['firsthalf_away_goals']
```

```python
secondhalf_conditions = [df_feature_eng['secondhalf_home_goals'] > df_feature_eng['secondhalf_away_goals'],
              df_feature_eng['secondhalf_home_goals'] < df_feature_eng['secondhalf_away_goals'],
              df_feature_eng['secondhalf_home_goals'] == df_feature_eng['secondhalf_away_goals']]

secondhalf_result = [df_feature_eng['home_team'],
          df_feature_eng['away_team'],
          'Draw']

df_feature_eng['secondhalf_result'] = np.select(secondhalf_conditions, secondhalf_result)
```
